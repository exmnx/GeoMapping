```python
#Importing Libraries

import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import folium
```

```python
#reading excel sheet
Jdata=pd.read_excel("/content/JapanOv.xlsx",sheet_name="Trip Overview",header=1)
Jdata.head()
```

```python
#Creating empty variable to populate new columns in data
temp= np.repeat(0,Jdata.shape[0])

#Creating new columns
Jdata["Latitude"]=temp
Jdata["Longitude"]=temp
Jdata['Latitude'].head()
```

```python
#Assgining geolocator
geolocator = Nominatim(user_agent="JdataAPP")
templat=[]
templist=[]
templong=[]
```

```python
#Using a for loop to get geolocations of all places in my data sheet using the column name of place
# and using list to store the locations
for x in range(0,len(Jdata['Name of Place'])):
    location = geolocator.geocode(Jdata['Name of Place'][x])
    templist.append(location)
```

```python
#Seeing if my list is the proper length, and seeing if the first location lat and long worked
print(len(templist))
print(templist[0].latitude,templist[0].longitude)
```

```python
#Creating for loop to store latitude and longitude in list for all places disregarding the places not found; therefore placign 0s instead

for x in range(0,len(templist)):
  try:
        templat.append(templist[x].latitude)
  except:
        templat.append(0)
```

```python
for x in range(0,len(templist)):
   try:
        templong.append(templist[x].longitude)
   except:
        templong.append(0)
```

```python
#Assigning values to dataframe columns
Jdata["Longitude"]=templong
Jdata["Latitude"]=templat
Jdata.head()
```

```python
#Rows in where there is no lat, long ; Assigning it a temporary array to use later
temparray=np.where(Jdata['Latitude']==0)
temparray
```

```python
#New Column in DF to look for lat,long of 0 values
Jdata['Nearby Alternative']=temp
```

```python
#This list lists nearby places or fixes the name of the place for the geolocator to work
#Manual input
temporaryList=['Sunshine City Japan','Semba Center','Round1 Osaka','Shitennoji','Osaka Wheel',
               'Osakajo Pier','Osaka Castle','Maguroya Shibata',
               'Hozenji temple Osaka','LUCUA Osaka','Namba Station','Hozenji temple Osaka',
               'Tanimachi 5,Osaka','Jinmyo Pass,Kyoto','Arashiyama Nakaoshitacho',
               'Honenji, kyoto','Hiei,kyoto','Chudoji Yabunouchicho','Kyoto-Kawaramachi Station','Daigo-ji',
               'Ninose station','Amanohashidate','Amanohashidate','Wakakusayama',
               'Mt. Kasuga, Nara','Nara Prefectural University']
```

```python
#Assigning the names above to the new column with proper row (Where 0's in lat,long)
for  x in temparray:
     Jdata['Nearby Alternative'][x]=temporaryList
     print(Jdata['Nearby Alternative'][x])
```

```python
templist1=[]
templat1=[]
templong1=[]
```

```python
#Finding geolocations of new names (missing lat,long rows)
#Storing in list
for x in range(0,len(temporaryList)):
    location1 = geolocator.geocode(temporaryList[x])
    templist1.append(location1)
```

```python
#Storing lat,long in temporary variables
for x in range(0,len(templist1)):
  try:
        templat1.append(templist1[x].latitude)
  except:
        templat1.append(0)

for x in range(0,len(templist1)):
   try:
        templong1.append(templist1[x].longitude)
   except:
        templong1.append(0)
```

```python
#Assigning new lat, long in columns of missing lat, long.
for x in temparray:
  Jdata["Longitude"][x]=templong1
  Jdata["Latitude"][x]=templat1
```

```python
#Checking for missing lat, long
np.where(Jdata['Latitude']==0)
```

```python
#Removing white spaces in Prefecture name
for x in range(0,len(Jdata['Prefectures'])):
  Jdata['Prefectures'][x]=Jdata['Prefectures'][x].strip(" ")
```

```python
#Creating temporary variables for each prefecture with specifc rows
varName=Jdata['Prefectures'].unique()
prefVar={}
k=0
while k<len(varName):
          var1=np.array(np.where(Jdata['Prefectures']==varName[k]))
          value=var1
          prefVar[k]=value
          k+=1
```

```python
prefVar
```

```python
#Creating temporary variables for each Ward with specifc rows
Jdata['Wards'].unique()
varName1=Jdata['Wards'].unique()
wardVar={}
k=0
while k<len(varName1):
          var2=np.array(np.where(Jdata['Wards']==varName1[k])[0][0])
          value=var2
          wardVar[k]=value
          k+=1
```

```python
#add in for loop and function to automatically add in the locations for the different prefects.

def mapFunction(length,prefec1):
    Mappy=folium.Map(location=(Jdata['Latitude'][length[0]], Jdata['Longitude'][length[0]]),zoom_start=12)

    for x in length:
          folium.Marker(
                location=[Jdata['Latitude'][x], Jdata['Longitude'][x]],
                #tooltip="Click here",
                #radius=2, weight=5,
                popup=Jdata["Name of Place"][x],
                #icon=folium.Icon(icon="bridge",prefix='fa'),
          ).add_to(Mappy)


#code for Circle Distancing in meters, to add miles change the number near miles
#Creates % mile circles for each ward in every prefecture map

    for x in range(0,len(Jdata['Wards'].unique())):
          if Jdata['Prefectures'][wardVar[x]]==prefec1:
              tempIndx=wardVar[x]
              miles=1
              radius = 1609.344*miles
              folium.Circle(
                    location=[Jdata['Latitude'][tempIndx], Jdata['Longitude'][tempIndx]],
                    radius=radius,
                    color="black",
                    weight=1,
                    fill_opacity=0.3,
                    opacity=1,
                    fill_color="pink",
                    fill=False,
                    popup="{} miles".format(miles),
              ).add_to(Mappy)


    return(Mappy)
```
