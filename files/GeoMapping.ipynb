```python
#Importing Libraries

import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import folium
```

```python
#reading excel sheet
Jdata=pd.read_excel("/content/JapanOv.xlsx",sheet_name="Trip Overview")
Jdata.head()
```

```python
#Creating empty variable to populate new columns in data
EmptyVar= np.repeat(0,Jdata.shape[0])

#Creating new columns
Jdata["Latitude"]=EmptyVar
Jdata["Longitude"]=EmptyVar
```

```python
#Assgining geolocator
geolocator = Nominatim(user_agent="JdataAPP")
templat=[]
templist=[]
templong=[]
```

```python
#Using a for loop to get geolocations of all places in my data sheet using the column name of place
# and using list to store the locations
for x in range(0,len(Jdata['Name of Place'])):
    location = geolocator.geocode(Jdata['Name of Place'][x])
    templist.append(location)
```

```python
#Seeing if my list is the proper length, and seeing if the first location lat and long worked
print('The length of this list is ' + str(len(templist)))
print('The dimesions of the dataframe is ' + str(np.shape(Jdata)))
print('The coordinates to the first location are',templist[0].latitude,templist[0].longitude)
```

```python
#Creating for loop to store latitude and longitude in list for all places disregarding the places not found; therefore placign 0s instead

for x in range(0,len(templist)):
  try:
        templat.append(templist[x].latitude)
  except:
        templat.append(0)
```

```python
for x in range(0,len(templist)):
   try:
        templong.append(templist[x].longitude)
   except:
        templong.append(0)
```

```python
#Assigning values to dataframe columns
Jdata["Longitude"]=templong
Jdata["Latitude"]=templat
Jdata.head()
```

```python
#Removing white spaces in Prefecture name
for x in range(0,len(Jdata['Prefectures'])):
  Jdata['Prefectures'][x]=Jdata['Prefectures'][x].strip(" ")
```

```python
#Creating temporary variables for each prefecture with specifc rows
varName=Jdata['Prefectures'].unique()
prefVar={}
k=0
while k<len(varName):
          var1=np.array(np.where(Jdata['Prefectures']==varName[k]))
          value=var1
          prefVar[k]=value
          k+=1
```

```python
#Creating temporary variables for each Ward with specifc rows
Jdata['Wards'].unique()
varName1=Jdata['Wards'].unique()
wardVar={}
k=0
while k<len(varName1):
          var2=np.array(np.where(Jdata['Wards']==varName1[k])[0][0])
          value=var2
          wardVar[k]=value
          k+=1
```

```python
#add in for loop and function to automatically add in the locations for the different prefects.

def mapFunction(length,prefec1,miles):
    Mappy=folium.Map(location=(Jdata['Latitude'][length[0]], Jdata['Longitude'][length[0]]),zoom_start=12)

    for x in length:
          folium.Marker(
                location=[Jdata['Latitude'][x], Jdata['Longitude'][x]],
                #tooltip="Click here",
                #radius=2, weight=5,
                #color="",
                popup=Jdata["Name of Place"][x],
                #icon=folium.Icon(icon="bridge",prefix='fa'),
          ).add_to(Mappy)


#code for Circle Distancing in meters, to add miles change the number of miles
#Creates % mile circles for each ward in every prefecture map

    for x in range(0,len(Jdata['Wards'].unique())):
          if Jdata['Prefectures'][wardVar[x]]==prefec1:
              tempIndx=wardVar[x]
              radius = 1609.344*miles
              folium.Circle(
                    location=[Jdata['Latitude'][tempIndx], Jdata['Longitude'][tempIndx]],
                    radius=radius,
                    color="black",
                    weight=1,
                    fill_opacity=0.3,
                    opacity=1,
                    fill_color="pink",
                    fill=False,
                    popup=str(Jdata['Wards'][wardVar[x]]) + ' Ward',
              ).add_to(Mappy)


    return(Mappy)
```

```python
'''
mapFunction(length,prefec1,miles); where length is the prefVar[0][0], [1][0], etc with respect
to Jdata['Prefectures'].unique(), prefec1 is the prefecture you want to look at, and miles is the
number of miles you want your ward circles to be.

example:
prefVar[0][0] = Osaka  ; prefVar[1][0] = Kyoto
prefVar[2][0] = Nara  ; prefVar[3][0] = Kanagawa
prefVar[4][0] = Tokyo ; prefVar[5][0] = Ibaraki
prefVar[6][0] = Yamanashi ;
'''
```
